```python
from selenium import webdriver               # Autoscrolling the entire webpage and extracting the html in soup object
from selenium.webdriver.common.action_chains import ActionChains
from bs4 import BeautifulSoup
import time

url = "https://raritysniper.com/nft-collections"

driver = webdriver.Chrome('/tmp/chromedriver')
driver.get(url)


last_height = driver.execute_script("return document.body.scrollHeight")
while True:
    
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(20)
    new_height = driver.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
soup = BeautifulSoup(driver.page_source, "html.parser")
# Scrape the loaded content

driver.quit()
```

```python
nft_links=[]  #finding all the links from the soup object
for link in soup.find_all('a'):
    href = link.get('href')
    #if href.startswith('https://www.raritysniper.com/nft/'):
    nft_links.append(href)
len(nft_links)
```

```python
extracted_links=nft_links[17:]  #removing the initial links 

def makelink(text):#convert the link to full http:// format
    newstr='https://raritysniper.com'+text   
    return newstr
final_list=list(map(makelink,extracted_links))
    
```

```python
lis=[]  #finding the name's of the collections from the image tag
img_tag = soup.find_all('img')
for img in img_tag:
    alt_text = img['alt']
    lis.append(alt_text)

print(lis)  
```

```python
list_name=lis[35:]   #manipulating the data to create a clean list of names
def filter_list(lst):
    counts = {}
    result = []
    for item in lst:
        if item not in counts:
            counts[item] = 1
        else:
            counts[item] += 1
    for item in counts:
        if counts[item] == 3:
            result.append(item)
    return result
result = filter_list(list_name)
print(result)
```
